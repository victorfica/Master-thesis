```python
import numpy as np

amino3to1dict = {'ASH': 'A','ALA': 'A','CYX': 'C','CYS': 'C','ASP': 'D','GLU': 'E','PHE': 'F','GLY': 'G','HIS': 'H',
                 'HID': 'H','HIE': 'H','HIP': 'H','ILE': 'I','LYS': 'K','LEU': 'L','MET': 'M','MSE': 'M','ASN': 'N',
                 'PYL': 'O','HYP': 'P','PRO': 'P','GLN': 'Q','ARG': 'R','SER': 'S','THR': 'T','SEL': 'U','VAL': 'V',
                 'TRP': 'W','TYR': 'Y'}
aa_std = "ACDEFGHIKLMNPQRSTVWY"

def dfpdb_res_level(tmp_df):
    """Obtener una visualizacion de pandas dataframe a nivel de residuo.
    Modificado de pandas_pdb.py method amino3to1 para funcionar con numeracion de residuos en anticuerpos"""

    cmp = 'placeholder'
    indices = []

    for num, ind in zip(tmp_df[['residue_number', 'insertion']].iterrows(), np.arange(tmp_df.shape[0])):
        numstr = '%s%s' % (num[1][0], num[1][1])
        if numstr != cmp:
            indices.append(ind)
        cmp = numstr
    tmp_df_res_level = tmp_df.iloc[indices].copy()
    return tmp_df_res_level

def contacts_by_aa(biopdb_df, contact_dataframe, protein="p1"):

    dfpdb_res = dfpdb_res_level(biopdb_df)
    print(dfpdb_res)
    # add col chain/resi for vectorize comparizon betwen contact res id and biopandas pdb
    dfpdb_res["chain/resi"] = dfpdb_res.chain_id.astype(str) + "/" + dfpdb_res.residue_number.astype(
        str) + dfpdb_res.insertion.astype(str)
    # group contact data by chain/resnum tag , created concatenating strings of atom
    if protein == "p1":
        key = contact_dataframe.atom_bgn.str.split("/", expand=True)[0].astype(str) + "/" + \
              contact_dataframe.atom_bgn.str.split("/", expand=True)[1].astype(str)
    else:
        key = contact_dataframe.atom_end.str.split("/", expand=True)[0].astype(str) + "/" + \
              contact_dataframe.atom_end.str.split("/", expand=True)[1].astype(str)
    contact_data_grouped = contact_dataframe.groupby(by=key).sum()
    contact_data_grouped.reset_index(inplace=True)

    # compare grouped data "chain/id"(index) with dfpdb_resn ("chain/resi") and get aminoacidname,add to contac_data
    oneletter_resname = dfpdb_res["residue_name"].loc[dfpdb_res["chain/resi"].isin(contact_data_grouped["index"])].map(
        amino3to1dict)
    contact_data_grouped["aminoacid"] = oneletter_resname.values

    # group contact_data by aminoacid name
    data_by_aa = contact_data_grouped.groupby("aminoacid").sum()

    # check not present aminoacids in data, and add resdue name with 15D array of zeros, finally sort indexes
    for aa in list(aa_std):
        if aa not in data_by_aa.index:
            data_by_aa.loc[aa] = np.zeros(15, dtype=int)
    data_by_aa.sort_index(inplace=True)

    return data_by_aa.stack().values
```
