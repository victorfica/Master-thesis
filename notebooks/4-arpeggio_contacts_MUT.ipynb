```python
import os
import pickle
import prody as pry
import sys
import subprocess
import warnings
import pandas as pd
from multiprocessing import Pool
import time
from pathlib import Path, PureWindowsPath

ABPRED_DIR = Path.cwd().parent
if ABPRED_DIR not in sys.path:
    sys.path.append(str(ABPRED_DIR))
    
from AbPred.interface_contacts import InterfaceContacts
from AbPred.features_ic import ArpeggioFeatures
```

```python
#Funcion para el sorting de paths por numero
def extract_num(pdb):
    return int(str(pdb.name).split(".")[2])

def sort_repair(pdb):
    return int(str(pdb.name).split(".")[2].split('_')[0])
```

```python
# Paths pdb data
PDBS_DIR = Path("../data/modelled_mutations/")
pdb_paths = list(PDBS_DIR.glob("*Repair.clean.pdb"))

# sorted
pdb_paths.sort(key=sort_repair)

# Data csv 
ab_bind_mCSM_HM = pd.read_csv("../data/ab_bind_mCSM_HM.csv",index_col=0)
```

```python
def main(i,pdbfile):
    
    partners_pdbs = ab_bind_mCSM_HM["Partners(A_B)"].iloc[i]
    ic = InterfaceContacts(pdbfile)
    print("### Running arpeggio ###")
    ic.run_arpeggio()
    time.sleep(10)
    ic.get_ic(partners=partners_pdbs)
    print("Finish getting interface contact\n")
    #ic.get_pymol_contacts()
    print("### END ###\n")
```

```python
with Pool(14) as p:
    print(p.starmap(main, enumerate(pdb_paths,0)))
```

# Contact analisys

```python
interface_sifts = []
pdbs_names = []
for paths in PDBS_DIR.glob("*.interface_contacts"):
    name = paths.name[:-19]
    contact_data = pd.read_table(paths)
    interface_sift_sum = contact_data.drop(["atom_bgn","atom_end","interacting_entities"],axis=1).sum()
    
    interface_sifts.append(interface_sift_sum.values)
    pdbs_names.append(name)
```

```python
col_names = contact_data.drop(["atom_bgn","atom_end","interacting_entities"],axis=1).columns
DF_features = pd.DataFrame(interface_sifts,columns=col_names,index=pdbs_names)
```
