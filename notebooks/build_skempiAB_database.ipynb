```python
from pathlib import Path
import sys
import pandas as pd
import shutil
import os
from multiprocessing import Pool


ABPRED_DIR = Path.cwd().parent
if ABPRED_DIR not in sys.path:
    sys.path.append(str(ABPRED_DIR))
from AbPred.FoldX import FoldX
```

```python
# Skempi dataframe of selected AB complexes
ab_data_singleMut = pd.read_csv('../data/skempi_ABlike_singleMut.csv',index_col=0)
ab_data_singleMut.sort_values('#Pdb',inplace=True)

# Paths data
PDBS_DIR = Path("../data/skempi2/")
skempi_pdbs = list(PDBS_DIR.glob("*.pdb"))
```

```python
ab_data_singleMut.shape
```

```python
ab_skempi_unique = ab_data_singleMut['#Pdb'].str.split('_',expand=True)[0].unique()
skempiAB_pdbs = list()

for pdb in skempi_pdbs:
    if pdb.name[:-4] in ab_skempi_unique:
        skempiAB_pdbs.append(pdb)
```

```python
# Copy selected pdbs to new dir
def copy_(source,target):
    
    target_path = Path(target)
    if not target_path.is_dir():
        target_path.mkdir()
    shutil.copy(source,target_path) 
    
for pdb in skempiAB_pdbs:
    copy_(pdb,'skempiAB')
```

# Repair pdbs

```python
def main(pdb):
    
    foldx = FoldX(verbose=True)
    foldx.repair_pdb(pdb_file=pdb)
```

```python
CWD = Path.cwd()

try:
    os.chdir('skempiAB')
    with Pool(12) as p:
        print(p.map(main, [str(pdb.name) for pdb in skempiAB_pdbs]))
        
finally:
    os.chdir(CWD)
```

```python
skempiAB_repairPDB = list(Path('skempiAB').glob("*Repair.pdb"))
```

# Generate mutants

```python
def mainPssm(row):
    # Extrtact relevant info from dataframe
    pdb_partners = row['#Pdb']
    pdbid = pdb_partners.split('_',maxsplit=1)[0]
    partners = pdb_partners.split('_',maxsplit=1)[1]    
    mutations_cleaned = row['Mutation(s)_cleaned']
    pdbname = '{}_Repair.pdb'.format(pdbid)
    
    print('Running foldx Pssm for',pdbname)
    foldx = FoldX(verbose=True)
    foldx.build_model(pdb_file=pdbname,foldx_mutation=mutations_cleaned)
```

```python
CWD = Path.cwd()
try:
    os.chdir('skempiAB')
    with Pool(12) as p:
        print(p.map(mainPssm, [row for i,row in ab_data_singleMut.iterrows()]))
        
finally:
    os.chdir(CWD)
```
