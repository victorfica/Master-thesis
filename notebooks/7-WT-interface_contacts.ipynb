```python
import pandas as pd
import numpy as np
import os
from pathlib import Path
import pickle
import sys

from multiprocessing import Pool
import time

ABPRED_DIR = Path.cwd().parent
if ABPRED_DIR not in sys.path:
    sys.path.append(str(ABPRED_DIR))
    
from AbPred.interface_contacts import InterfaceContacts
```

```python
#Funcion para el sorting de paths por numero
def extract_num(pdb):
    return int(pdb.name.split(".")[2])
```

```python
# Paths
DATA = ABPRED_DIR / "data"
PDBS_DIR = Path("wt_modells/")

ab_bind_mCSM_HM = pd.read_csv(DATA / "ab_bind_mCSM_HM.csv")
```

```python
wt_pdbs = list(PDBS_DIR.glob("*.pdb"))
wt_pdbs.sort(key=extract_num)
```

```python
def main(i,pdbfile):
    pdbfile = str(pdbfile.absolute())
    partners_pdbs = ab_bind_mCSM_HM["Partners(A_B)"][i]
    
    ic = InterfaceContacts(pdbfile)
    print("### Running arpeggio ###")
    ic.run_arpeggio()
    time.sleep(15)
    ic.get_ic(partners=partners_pdbs)
    print("Finish getting interface contact\n")
    ic.get_pymol_contacts()
    print("### END ###\n")
    
```

```python
with Pool(12) as p:
    print(p.starmap(main, enumerate(wt_pdbs)))
```
