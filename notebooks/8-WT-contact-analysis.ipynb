```python
import pandas as pd
import os
from pathlib import Path
import pickle
```

```python
ABPRED_DIR = Path(os.getcwd()).parent
DATA = ABPRED_DIR / "data"
PDBS_WT = Path("wt_modells")
```

```python
PDBS_WT.is_dir()
```

```python
interface_sifts = []
pdbs_names = []
for paths in PDBS_WT.glob("*.interface_contacts"):
    name = paths.name[:-19]
    contact_data = pd.read_table(paths)
    interface_sift_sum = contact_data.drop(["atom_bgn","atom_end","interacting_entities"],axis=1).sum()
    
    interface_sifts.append(interface_sift_sum.values)
    pdbs_names.append(name)
```

```python
col_names = contact_data.drop(["atom_bgn","atom_end","interacting_entities"],axis=1).columns
DF_features = pd.DataFrame(interface_sifts,columns=col_names,index=pdbs_names)
DF_features.to_csv(DATA / "DF_wildtype_features.csv")
```

```python
DF_features.covalent.value_counts()
```

```python
pdbs_with_cov = DF_features.loc[DF_features["covalent"] != 0].index.tolist()
```

```python
pdb2repair = []
for pdb in pdbs_with_cov:
    for path in PDBS_WT.glob("*.pdb"):
        if pdb in str(path):
            pdb2repair.append(str(path.absolute()))
```

```python
pickle.dump(pdb2repair,open("pdbsWT_to_repair.pkl","wb"))
```
